```python
import os
import time
import imageio
import numpy as np
import random as rn
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def Graph_and_Coor(n):
    
    G = nx.Graph() # create an empty graph
    n2 = n**2 # n x n
    
    G.add_nodes_from([i for i in range(1,n2+1)]) # add n^2 nodes to the Graph
        
    Coordinates = {} 
    node = 1
    
    for i in range(n,0,-1): # for chosen row, (3,2,1) reverse "for" - last row is on y = 1, interval = 1
        for j in range(1,n+1): # and chosen column
            Coordinates[node] = (j,i) #(1,3),(2,3),(3,3),...,(2,1),(3,1)
            node = node + 1 
            
    # Coordinates for 3x3 {1: (1, 3), 2: (2, 3), 3: (3, 3), 4: (1, 2), 5: (2, 2), 6: (3, 2), 7: (1, 1), 8: (2, 1), 9: (3, 1)}
    
    nodes_list = list(G.nodes())

    for i in nodes_list: ### add edges
        if i%n != 0:
            G.add_edge(i,i+1, weight=1) #(1,2),(2,3),(3,4),...
    for i in nodes_list:
        if i <= n2-n:
             G.add_edge(i,i+n, weight=1) #(1,6),(2,7),(3,8),...
        else:
            break
    for i in nodes_list[:-n-1]: #vertices without last row (1 on y axis)
        if i%n != 0:
             G.add_edge(i,i+n+1,weight=1) #diagonal down edges 
            
    for i in nodes_list[n:]: #vertices without first row (n on y axis)
        if i%n != 0:
             G.add_edge(i,i-n+1,weight=1) #diagonal up edges 
                
    return G,Coordinates
```

```python
def Graph_and_Coor8(n):
    
    G = nx.Graph() # create an empty graph
    n2 = n**2 # n x n
    
    G.add_nodes_from([i for i in range(1,n2+1)]) # add n^2 nodes to the Graph
        
    Coordinates = {} 
    node = 1
    
    for i in range(n,0,-1): # for chosen row, (3,2,1) reverse "for" - last row is on y = 1, interval = 1
        for j in range(1,n+1): # and chosen column
            Coordinates[node] = (j,i) #(1,3),(2,3),(3,3),...,(2,1),(3,1)
            node = node + 1 
            
    # Coordinates for 3x3 {1: (1, 3), 2: (2, 3), 3: (3, 3), 4: (1, 2), 5: (2, 2), 6: (3, 2), 7: (1, 1), 8: (2, 1), 9: (3, 1)}
    
    nodes_list = list(G.nodes())

    for i in nodes_list: ### add edges
        if i%n != 0:
            G.add_edge(i,i+1, weight=1) #(1,2),(2,3),(3,4),...
    for i in nodes_list:
        if i <= n2-n:
             G.add_edge(i,i+n, weight=1) #(1,6),(2,7),(3,8),...
        else:
            break
    for i in nodes_list[:-n]: #vertices without last row (1 on y axis)
        if i%n != 0: #if rest from div
             G.add_edge(i,i+n+1,weight=1) #diagonal down edges 
            
    for i in nodes_list[n:]: #vertices without first row (n on y axis)
        if i%n != 0:
             G.add_edge(i,i-n+1,weight=1) #diagonal up edges 
                
    for i in nodes_list:
        if i == n:
            G.add_edge(i,n2-1,weight=1)
            G.add_edge(i,n2,weight=1)
            G.add_edge(i,n2-n+1,weight=1)
            G.add_edge(i,1,weight=1)
            G.add_edge(i,n+1,weight=1)
        elif i%n == 0 and i>n and i<n2:
            G.add_edge(i,i-n+1,weight=1)
            G.add_edge(i,i+1,weight=1)
            G.add_edge(i,i-2*n+1,weight=1)
        elif i == n2:
            G.add_edge(i,n2-2*n+1,weight=1)
            G.add_edge(i,n2-n+1,weight=1)
            G.add_edge(i,1,weight=1)
            G.add_edge(i,n-1,weight=1)
        elif i>1 and i<n:
            G.add_edge(i,i+n2-n,weight=1)
            G.add_edge(i,i+n2-n+1,weight=1)
            G.add_edge(i,i+n2-n-1,weight=1)
            G.add_edge
        elif i == 1:
            G.add_edge(i,i+n2-n,weight=1)
            G.add_edge(i,i+n2-n+1,weight=1)
            
    return G,Coordinates
```

```python
def Conway_game_with_offset(file_name,n,markersize,CreateGIF,draw,offset):
    
    n2 = n**2
 
    G,Coordinates = Graph_and_Coor(n)
           
    States_dict = {} # DICTIONARY, FOR EACH VERTEX WE HAVE 'dead' or 'alive'
    
    txt_file = np.genfromtxt(file_name,delimiter=',') # read txt file
    i = 1
    for line in txt_file: # take line by line
        for element in line: # for element in line check if it is 0 or 1
            if element == 1:
                States_dict[i] = 'alive'
                i = i+1
            else:
                States_dict[i] = 'dead'
                i = i+1         

    ###############################################################        
    if CreateGIF == 1: # if we want to create GIF
        cwd = os.getcwd() # DIRECTORY TO SAVE FIGRURES
        where = os.path.join(cwd+"\\","Steps")
        if not os.path.exists(where):
            os.mkdir(where)
        
    if draw == 1:
        
        plt.figure() 
        
        for i in range(offset*n+1,n2+1-offset*n):
            if i%n != 1 and i%n != 2 and i%n != n-1 and i%n != 0:
                state = States_dict[i] 
                x = Coordinates.get(i)[0] 
                y = Coordinates.get(i)[1]
            
                if state == 'alive':
                    plt.plot(x,y,'gs',markersize = markersize)
                else:
                    pass

                
    if CreateGIF == 1:
        file = os.path.join(where,"Life"+str(1))
        plt.axis('off')
        plt.plot([0,0,n+1,n+1],[0,n+1,0,n+1],'ks',markersize = markersize)
        plt.title('Game of Life'+str(n))
        plt.savefig(file + '.png')
        plt.close()
    #################################################################
    
    go = 1 
    steps = 1
    
    # next steps 
    while steps < 500 and go == 1: #NEXT STEPS
        
        steps = steps+1
        States_dict_new = States_dict.copy() # UPDATE WITH RESPECT TO ONE STEP BEFORE 
        nodes_list = list(G.nodes())
        
        nodes_to_draw = []
        static_nodes = []
        for i in nodes_list:
            
            if States_dict[i] == "dead":
                
                ALIVE = 0
                for nei in G.neighbors(i):
                    if States_dict[nei] == 'alive':
                        ALIVE = ALIVE +1
                if ALIVE == 3:
                    States_dict_new[i] = "alive"
                    nodes_to_draw.append(i)
                    
            elif States_dict[i] == "alive":
                
                ALIVE = 0
                for nei in G.neighbors(i):
                    if States_dict[nei] == 'alive':
                        ALIVE = ALIVE +1
                        
                if ALIVE != 2 and ALIVE != 3:
                    States_dict_new[i] = "dead"
                    nodes_to_draw.append(i)
                else:
                    static_nodes.append(i)
                    nodes_to_draw.append(i)
                    
        if len(nodes_to_draw)-len(static_nodes) == 0:
            go = 0
        #print(len(nodes_to_draw)-len(static_nodes))
            
        States_dict = States_dict_new
        
    ##################################################################        
        if draw == 1:
            
            plt.figure()
            
            for i in range(offset*n+1,n2+1-offset*n):
                if i%n != 1 and i%n != 2 and i%n != n-1 and i%n != 0:
                    if i in nodes_to_draw:

                        state = States_dict[i] 
                        x = Coordinates.get(i)[0] 
                        y = Coordinates.get(i)[1]

                        if state == 'alive':
                            plt.plot(x,y,'gs',markersize = markersize)
                        else:
                            pass
           

        if CreateGIF == 1:
            file = os.path.join(where,"Life"+str(steps))
            plt.axis('off')
            plt.plot([0,0,n+1,n+1],[0,n+1,0,n+1],'ks',markersize = markersize)
            plt.title('Game of Life'+str(n))
            plt.savefig(file + '.png')
            plt.close()
    #################################################################
    
    if CreateGIF == 1:
        ### open gif file
        with imageio.get_writer(cwd+'\\Game_of_Life.gif', mode='I',fps = 12) as writer:
            
            for root, dirs, files in os.walk(cwd +"\\Steps"): # find all files using in GIF

                for i in range(len(files)):
                    files[i] = cwd +"\\Steps\\" + files[i]
                files.sort(key=os.path.getctime) # sort files by file create time
                
                for file in files: # create a GIF
                    image = imageio.imread(file)
                    writer.append_data(image)
                writer.close()
                
```
